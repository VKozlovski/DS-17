Напишіть функції, що:
* рохраховує мінімальне та максимальне значення
* середнє значення послідовності
* медіану послідовності
* частоту кожної цифри в послідовності
* середнє квадратичне відхилення

Послідовність чисел ганеруємо рандомно

```python
import random

#генеруэмо тисячу чисел в діапазоні від 0 до 10
list_with_random_numbers = [random.randint(0,100) for i in range(1000)]

                    
```

```python

import statistics
print(f'Максимальне значення = {max(list_with_random_numbers)}')
print(f'Мінімальне значення = {min(list_with_random_numbers)}')
print(f'Середнє значення = {sum(list_with_random_numbers)/len(list_with_random_numbers)}')
print(f'Медіана послідовності = {statistics.median(list_with_random_numbers)}')
```

```python
dictOfNumbers = {}
for i in list_with_random_numbers:
    if i in dictOfNumbers:
        dictOfNumbers[i] += 1
    else:
        dictOfNumbers[i] = 1

print(f'Словник входжень:\n{dictOfNumbers}')
```

```python
import numpy
print(f'Середнє квадратичне відхилення: {numpy.std(list_with_random_numbers)}')
```
